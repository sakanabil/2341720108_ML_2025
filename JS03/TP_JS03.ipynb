```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python

from google.colab import drive
drive.mount('/content/drive')

file_path = "/content/drive/My Drive/Colab Notebooks/wbc.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
df = df.drop(columns=['id'])
```

```python
df.head()
```

```python
# 2. Encoding diagnosis (M = 1, B = 0)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.head()
```

```python
# 3. Standardisasi kolom numerik
kolom_numerik = df.drop(columns=['diagnosis']).columns
scaler = StandardScaler()
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Stratified split 80:20
df_train, df_test = train_test_split(
    df,
    test_size=0.2,
    random_state=0,
    stratify=df['diagnosis']
)

# Cek distribusi label
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
